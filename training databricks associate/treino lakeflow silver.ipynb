```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DateType, TimestampType
import pyspark.sql.functions as F
```

```python
env = "dev"
schema = "silver"
table_male = "covid_male"
table_female = "covid_female"
table_name_male = f"{env}.{schema}.{table_male}"
table_name_female = f"{env}.{schema}.{table_female}"
```

```python
df_silver = spark.table("dev.bronze.covid")
df_silver.count()
```

```python
display(df_silver.limit(10))
```

```python
df_female = (
    df_silver.filter(F.col("sex") == "female")
)
df_male = (
    df_silver.filter(F.col("sex") == "male")
)
print("qtd female: ",df_female.count())
print("qtd male: ",df_male.count())
```

```python
def write_delta_table(df, table_name):
  (df.write
   .mode("overwrite")
   .format("delta")
   .saveAsTable(table_name)
  )
```

```python
spark.sql(f"CREATE CATALOG IF NOT EXISTS {env}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {env}.{schema}")
```

```python
write_delta_table(df_male,table_name_male)
write_delta_table(df_female,table_name_female)
```
